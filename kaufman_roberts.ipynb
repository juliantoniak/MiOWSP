```python
def calc_x(V, M, a, t): #obliczanie xn
    x = [1] * (V + 1)
    for n in range(1, V+1):
        sum = 0
        for i in range(0, M):
            if n >= t[i]:
                sum += a[i] * t[i] * x[n - t[i]]
        x[n] = sum / n
    return x

def calc_p0(x): #wartość P(0)
    sum = 0
    for item in x:
        sum += item
    return 1 / sum

def calc_pn(x, V, M, a, t): #wartość P(n)
    P = [1] * (V + 1)
    P[0] = calc_p0(x)
    for n in range(1, V+1):
        sum = 0
        for i in range(0, M):
            if n >= t[i]:
                sum += a[i] * t[i] * P[n - t[i]]
        P[n] = sum / n
    return P


def calc_bn(P, V, t, i=1): #prawdopodobieństwo blokady strumienia zgłoszeń klasy n
    sum = 0
    for n in range(V - t[i -1] + 1, V + 1):
        sum += P[n]
    return sum

def calc_all(M, V, t, a): #Obliczanie
    x = calc_x(V, M, a, t)
    P = calc_pn(x, V, M, a, t)

    b = [1] * M
    i = 0
    while i < M:
        b[i] = calc_bn(P, V, t, i + 1)
        i += 1

    print("x = ", x)
    print("P = ", P)

    ii = 0
    while ii < M:
        print("b", ii+1, " = ", b[ii])
        ii += 1

M = 2 #strumienie ruchu (M = len(t[x]) = len(a[x]))
V = 3 #pojemność wiązki wyrażona w kanałach lub podstawowych jednostkach pasma V >= M
t = [[1, 2], [3, 4]] #liczba żądanych jednostek przetwarzania
a = [[0.4, 1], [1, 2]] #ruch oferowany

print("M = ", M)
print("V = ", V)
print("\n")

for itemA in a:
    for itemT in t:
        print("---" * 15)
        print("ai = ", itemA)
        print("ti = ", itemT)
        calc_all(M, V, itemT, itemA)
        print("---" * 15)
        print("\n")
```
